```python
df_country=df[df['location'] == 'India']
df_country
```

```python
df_country.fillna(0,inplace=True)
```

```python
df_country.head()
```

```python
df_country['date'] = pd.to_datetime(df_country['date'], format='%d-%m-%Y', errors='raise')
```

```python
df_country.dtypes
```

```python
df_country['daily growth rate']=(df_country['new_cases']/df_country['total_cases'])*100
```

```python
import matplotlib.pyplot as plt
# plt.plot(df_country['daily growth rate'])
plt.plot(df_country['date'], df_country['daily growth rate'], label='Daily Growth Rate', color='green')
plt.xlabel('Date')
plt.ylabel('Growth Rate (%)')
plt.title('COVID-19 Daily Growth Rate Over Time')
plt.legend()
plt.show()
```

```python
df_country['daily death rate']=(df_country['total_deaths']/df_country['total_cases'])*100
```

```python
plt.plot(df_country['date'],df_country['daily death rate'],label='Daily Death Rate',color='black')
plt.xlabel('Date')
plt.ylabel('Death Rate (%)')
plt.title('COVID-19 Daily Death Rate Over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 5))
plt.plot(df_country['date'], df_country['total_cases'], label='Total Cases')
plt.plot(df_country['date'], df_country['total_deaths'], label='Total Deaths', color='red')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('COVID-19 Cases and Deaths Over Time (India)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_country.columns
```

```python
import seaborn as sns
sns.scatterplot(x='total_cases',y='total_deaths',data=df_country,color='green')
plt.xlabel("Total cases identified")
plt.ylabel("Total deaths")
plt.title("Total cases vs Total deaths")
plt.show()
```

```python

sns.scatterplot(x=df_country['daily growth rate'], y=df_country['daily death rate'], color='purple')
plt.xlabel('Daily Growth Rate (%)')
plt.ylabel('Daily Death Rate (%)')
plt.title('Comparison of Daily Growth Rate and Daily Death Rate')
plt.show()
```

```python

cols_of_interest = ['daily growth rate', 'daily death rate', 'total_cases', 'total_deaths', 'total_tests']
plt.figure(figsize=(10, 8))
sns.heatmap(df_country[cols_of_interest].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of COVID-19 and Health Metrics')
plt.show()
```

```python
plt.figure(figsize=[10,5])
fig,axes=plt.subplots(1,3)
sns.barplot(x=df_country['aged_65_older'], y=df_country['daily death rate'], color='green')

sns.barplot(x=df_country['median_age'], y=df_country['daily death rate'], color='blue', ax=axes[0])
axes[0].set_title(' Median Age')
axes[0].set_xlabel('Median Age')
axes[0].set_ylabel('Daily Death Rate (%)')

sns.barplot(x=df_country['aged_65_older'], y=df_country['daily death rate'], color='green', ax=axes[1])
axes[1].set_title('Aged 65+')
axes[1].set_xlabel('Aged 65 or Older (%)')
axes[1].set_ylabel('Daily Death Rate (%)')

sns.barplot(x=df_country['aged_70_older'], y=df_country['daily death rate'], color='red', ax=axes[2])
axes[2].set_title('Aged 70+')
axes[2].set_xlabel('Aged 70 or Older (%)')
axes[2].set_ylabel('Daily Death Rate (%)')


plt.tight_layout()

plt.show()
```
